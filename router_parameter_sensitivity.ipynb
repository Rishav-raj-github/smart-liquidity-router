# Router Sensitivity Analysis
Analyzes how order allocation shifts as venue execution fees change relative to book spreads.


```python
import numpy as np
import matplotlib.pyplot as plt

fees = np.linspace(0.00, 0.10, 50)
ask_a = 100.05
ask_b = 100.10

# Cost to buy at Venue A vs Venue B depending on Venue A's fee
cost_a = ask_a + fees
cost_b = ask_b + 0.02

plt.figure(figsize=(10, 4))
plt.plot(fees, cost_a, label="Venue A Total Cost", color="green")
plt.axhline(cost_b, color="red", linestyle="--", label="Venue B Constant Cost ($100.12)")
plt.title("Routing Shift Threshold Analysis")
plt.xlabel("Venue A Execution Fee ($)")
plt.ylabel("Total Unit Cost ($)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

```
