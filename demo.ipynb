# Smart Liquidity Router Simulation
This notebook evaluates order splits across multiple execution venues to minimize the Volume-Weighted Average Price (VWAP) including venue execution fees.


```python
# Implement the router algorithm in Python to plot allocations
import matplotlib.pyplot as plt
import numpy as np

pools = [
    {"id": "Venue-A", "best_ask": 100.10, "size": 200, "fee": 0.02},
    {"id": "Venue-B", "best_ask": 100.05, "size": 100, "fee": 0.05},
    {"id": "Venue-C", "best_ask": 100.15, "size": 400, "fee": 0.01},
]

def route_order(qty):
    sorted_pools = sorted(pools, key=lambda x: x["best_ask"] + x["fee"])
    remaining = qty
    allocations = {}
    
    for p in sorted_pools:
        if remaining <= 0: break
        fill = min(remaining, p["size"])
        allocations[p["id"]] = fill
        remaining -= fill
        
    if remaining > 0:
        allocations["Slippage"] = remaining
    return allocations

# Test various order sizes
sizes = [50, 150, 300, 500, 800]
results = [route_order(s) for s in sizes]

for s, res in zip(sizes, results):
    print(f"Order Size: {s:3d} | Allocations: {res}")

```
