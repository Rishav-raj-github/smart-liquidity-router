# Multi-Venue Spread Arbitrage Simulation
Simulates pricing discrepancies between venues and computes transaction sizes to capture spread profits.


```python
price_venue_a = 100.05
price_venue_b = 100.15
fee = 0.02

spread = price_venue_b - price_venue_a
net_arbitrage_profit = spread - (fee * 2)

print(f"Venue A Ask: ${price_venue_a} | Venue B Bid: ${price_venue_b}")
print(f"Raw Spread: ${spread:.2f} | Net Arbitrage Spread: ${net_arbitrage_profit:.2f}")

```
